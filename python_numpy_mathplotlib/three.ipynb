```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0,2,100)
y = x ** 2

plt.figure(figsize=(5,5))
plt.plot(x,y,label="square")
plt.plot(x,x,label="linear")
plt.title("simple plot")
plt.legend()
plt.show()
```
